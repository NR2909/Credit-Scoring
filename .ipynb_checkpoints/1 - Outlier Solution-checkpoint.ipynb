# Data Preprocessing - Outlier detection

Data Set HMEQ
The data set HMEQ reports characteristics and delinquency information for 5,960 home equity loans. A home equity loan is a loan where the obligor uses the equity of his or her home as the underlying collateral. The data set has the following characteristics:
- BAD: 1 = applicant defaulted on loan or seriously delinquent; 0 = applicant paid loan
- LOAN: Amount of the loan request
- MORTDUE: Amount due on existing mortgage
- VALUE: Value of current property
- REASON: DebtCon = debt consolidation; HomeImp = home improvement
- JOB: Occupational categories
- YOJ: Years at present job
- DEROG: Number of major derogatory reports
- DELINQ: Number of delinquent credit lines
- CLAGE: Age of oldest credit line in months
- NINQ: Number of recent credit inquiries
- CLNO: Number of credit lines
- DEBTINC: Debt-to-income ratio

# Exercise 3.1

```python
# copy df
hmeq_iqr = hmeq.copy()
```

```python
from scipy import stats

cols = ["CLAGE", "DEBTINC"]

get_iqr = lambda x: {"lower":x.quantile(0.25) - 1.5 * stats.iqr(x), "upper":x.quantile(0.75) + 1.5 * stats.iqr(x)}

# compute iqr
iqr = hmeq_iqr[cols].apply(get_iqr)

# remove outliers
for col, bounds in zip(cols, iqr):
    rows_before = hmeq_iqr.shape[0]
    hmeq_iqr = hmeq_iqr[(hmeq_iqr[col] < bounds["upper"]) & (bounds["lower"] < hmeq_iqr[col])].reset_index(drop=True)
    print(f"{col}: removed {rows_before-hmeq_iqr.shape[0]} rows ...")
```

# Exercise 3.2

```python
from scipy import stats

cols = ["CLAGE", "DEBTINC"]

# compute zscores
zscores = hmeq[cols].apply(stats.zscore)

# zscores greater or equal 3 are set to False
mask = np.all(np.abs(zscores.values) < 3, axis=1)

# remove masked rows
hmeq_zscore = hmeq.iloc[mask].reset_index(drop=True)

print(f"Removed {hmeq.shape[0]-hmeq_zscore.shape[0]} rows ...")
```

```python
# CLAGE: 20 rows contain zscore >= 3
# DEBTINC: 24 rows contain zscore >= 3

(zscores > 3).sum()
```

# Exercise 3.3

```python
# copy df
hmeq_winsorized = hmeq.copy()
```

```python
from scipy import stats

cols = ["CLAGE", "DEBTINC"]

# 5% winsorization (2-sided -> 2.5% bottom, 97.5% top)
perc = 0.05

# winsorize column
hmeq_winsorized[cols] = hmeq_winsorized[cols].apply(stats.mstats.winsorize, limits=(perc/2, perc/2))
```

# Exercise 4

```python
# Compare the distribution of values for the original data, zscore treatment and winsorization treatment (for both CLAGE and DEBTINC)

for col in ["CLAGE", "DEBTINC"]:
    data = pd.concat([hmeq_iqr[col], hmeq_zscore[col], hmeq_winsorized[col]], axis=1)
    data.columns = ["iqr", "zscore", "winsorized"]

    fig, ax = plt.subplots(1,2, figsize=(12,3))
    plt.title(col)
    sns.kdeplot(hmeq[col], ax=ax[0])
    sns.kdeplot(data=data.melt(), x="value", hue="variable", alpha=0.5, bw_adjust=0.3, ax=ax[1])
    fig.show()
```
